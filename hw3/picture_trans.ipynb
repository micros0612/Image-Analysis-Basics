```python
from torchvision import transforms
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ori_pic=Image.open('njust.jpg')
display(ori_pic)
```

```python
#平移
trans=transforms.RandomAffine(degrees=0,translate=(0.4,0.2))(ori_pic)
#旋转
rot=transforms.RandomAffine(degrees=(22,38))(ori_pic)
#水平翻转
flip1=transforms.RandomHorizontalFlip(p=1)(ori_pic)
#垂直翻转
flip2=transforms.RandomVerticalFlip(p=1)(ori_pic)
```

```python
#显示
plt.figure(figsize=(22, 22))
plt.subplot(151), plt.imshow(ori_pic), plt.title("原图"), plt.xticks([]), plt.yticks([])
plt.subplot(152), plt.imshow(trans), plt.title("平移"), plt.xticks([]), plt.yticks([])
plt.subplot(153), plt.imshow(rot), plt.title("旋转"), plt.xticks([]), plt.yticks([])
plt.subplot(154), plt.imshow(flip1), plt.title("水平镜像"), plt.xticks([]), plt.yticks([])
plt.subplot(155), plt.imshow(flip2), plt.title("垂直镜像"), plt.xticks([]), plt.yticks([])
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.tight_layout(), plt.show()
```
